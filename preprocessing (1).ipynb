## **Normalization**

Here ,  we project image data pixels  to a predefined range  (0,1) or (-1, 1).

This also helps us to provide a common learning rate as we know high intensity pixels require low learning rate and low intensity pixels need a high learning rate. Normalizing them in a range gives advantage of providing standard learning rate.

```python
i, (im1, im2, im3, im4) = plt.subplots(1, 4, sharey=True)
i.set_figwidth(20) 
image=cv2.imread("/content/3.jpg")
gray_image = skimage.color.rgb2gray(image)
norm_image1 = (gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image))
image=cv2.imread("/content/1.jpg")
gray_image = skimage.color.rgb2gray(image)
norm_image2 = (gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image))
image=cv2.imread("/content/2.jpg")
gray_image = skimage.color.rgb2gray(image)
norm_image3 = (gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image))
image=cv2.imread("/content/6.jpg")
gray_image = skimage.color.rgb2gray(image)
norm_image4 = (gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image))

im1.imshow(norm_image1)   
im2.imshow(norm_image2)
im3.imshow(norm_image3)  
im4.imshow(norm_image4)
```

#Varying the image brightness

```python
from numpy import expand_dims
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
# convert to numpy array
data = np.array(image)

# expand dimension to one sample
samples = expand_dims(image, 0)

# create image data augmentation generator
datagen = ImageDataGenerator(brightness_range=[0.5,2.0])

# create an iterator
it = datagen.flow(samples, batch_size=1)
fig, im = plt.subplots(nrows=1, ncols=3, figsize=(15,15))

# generate batch of images
for i in range(3):

    # convert to unsigned integers
    image = next(it)[0].astype('uint8')
 
    # plot image
    im[i].imshow(image) 
```
